# Pokemon Analysis and Visualizations

```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
pokedata = pd.read_csv('Pokemon.csv')
```

```python
pokedata.head()
```

```python
iofhighesttotal = pokedata['Total'] == pokedata['Total'].max()
ioflowesttotal = pokedata['Total'] == pokedata['Total'].min()
```

```python
pokedata[iofhighesttotal]
```

```python
pokedata[ioflowesttotal]
```

```python
sns.countplot(data = pokedata, x = 'Type 1')
```

```python
pokedata['Type 1'].unique()
```

```python
#create dataframes of the different types of pokemon (based on primary types)
grass_pokemon  = pokedata[pokedata['Type 1'] == 'Grass']
fire_pokemon   = pokedata[pokedata['Type 1'] == 'Fire']
water_pokemon  = pokedata[pokedata['Type 1'] == 'Water']
bug_pokemon    = pokedata[pokedata['Type 1'] == 'Bug']
normal_pokemon = pokedata[pokedata['Type 1'] == 'Normal']
poison_pokemon  = pokedata[pokedata['Type 1'] == 'Poison']
electric_pokemon  = pokedata[pokedata['Type 1'] == 'Electric']
ground_pokemon  = pokedata[pokedata['Type 1'] == 'Ground']
```

```python
#creating a dataframe of legendary pokemon only
legendary_pokemon = pokedata[pokedata['Legendary'] == True]
legendary_pokemon
```

```python
#creating a dataframe of non-legendary pokemon only
reg_pokemon = pokedata[pokedata['Legendary'] == False]
reg_pokemon.head()
```

```python
#finding the non-legendary with the highest attack
highest_atk = reg_pokemon['Attack'] == reg_pokemon['Attack'].max()
reg_pokemon[highest_atk]
```

```python
#finding the non-legendary with the highest defense
highest_def = reg_pokemon['Defense'] == reg_pokemon['Defense'].max()
reg_pokemon[highest_def]
```

```python
#finding the non-legendary with the highest special attack
highest_spatk = reg_pokemon['Sp. Atk'] == reg_pokemon['Sp. Atk'].max()
reg_pokemon[highest_spatk]
```

```python
#finding the non-legendary with the highest special defense
highest_spdef = reg_pokemon['Sp. Def'] == reg_pokemon['Sp. Def'].max()
reg_pokemon[highest_spdef]
```

```python
#Plotting all pokemon with an atk of over 150
strongatk = pokedata[pokedata['Attack'] > 150 ]
sns.lmplot(data = strongatk,
           x = 'Defense',
           y = 'Attack',
           hue = 'Name',
           fit_reg = False)
```

```python
#Plotting all pokemon with an atk of over 150
strongatk = pokedata[pokedata['Sp. Atk'] > 150 ]
sns.lmplot(data = strongatk,
           x = 'Defense',
           y = 'Sp. Atk',
           hue = 'Name',
           fit_reg = False)
```

```python
strong_overall = pokedata[pokedata['Total']>600]
sns.lmplot(data = strong_overall,
           x = 'Defense',
           y = 'Attack',
           hue = 'Name',
           fit_reg = False)
```
